```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import cv2
import numpy as np

%matplotlib inline

#read the image and display it
image=img.imread('images/car_green_screen2.jpg')
plt.imshow(image)

#convert to HSV
hsv_image = cv2.cvtColor(image,cv2.COLOR_RGB2HSV)

#display the impage
plt.imshow(hsv_image)

#define upper and lower boundaries for green color in HSV
lower_green=np.array([10,66,23])
upper_green = np.array([102,255,255])

#create copy of original HSV image
copy_image = np.copy(hsv_image)

#create mask using set boundaries
mask_using_hsv=cv2.inRange(copy_image,lower_green,upper_green)

#create copy of original image to apply mask on
masked_image= np.copy(image)

#apply the mask
masked_image[mask_using_hsv != 0] = [0,0,0]

#display masked image
plt.imshow(masked_image)

#uncomment below line to see the original image which was being masked
#plt.imshow(image)
```
